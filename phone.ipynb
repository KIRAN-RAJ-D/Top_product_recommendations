```python
import pandas as pd
import re

# Load the CSV file into a DataFrame
df = pd.read_csv('merged_mobile_data.csv')

# Function to remove unwanted emojis
def remove_emojis(text):
    emoji_pattern = re.compile("[\U0001F600-\U0001F64F"  # emoticons
                                 "\U0001F300-\U0001F5FF"  # symbols & pictographs
                                 "\U0001F680-\U0001F6FF"  # transport & map symbols
                                 "\U0001F700-\U0001F77F"  # alchemical symbols
                                 "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                                 "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                                 "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                                 "\U0001FA00-\U0001FA6F"  # Chess Symbols
                                 "\U00002702-\U000027B0"  # Dingbats
                                 "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

# Check if 'review' column exists in the DataFrame
if 'review' in df.columns:
    # Apply the function to the 'review' column
    df['review'] = df['review'].apply(remove_emojis)

    # Optionally, save the cleaned DataFrame to a new CSV file
    df.to_csv('cleaned_mobile_data.csv', index=False)
    print("Emojis removed and cleaned data saved to 'cleaned_mobile_data.csv'.")
else:
    print("The 'review' column does not exist in the dataset.")

# Display the updated DataFrame
print(df.head())  # Display the first few rows to verify
```

```python
import pandas as pd
from textblob import TextBlob

# Load the data
df = pd.read_csv('mobile_data_reviews_cleaned.csv')

# Function to get sentiment
def get_sentiment(review):
    analysis = TextBlob(review)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the function to the review column
df['Sentiment'] = df['Review'].apply(get_sentiment)

# Count the occurrences of each sentiment
sentiment_counts = df['Sentiment'].value_counts()

# Display results
print(df[['Review', 'Sentiment']].head())
print(sentiment_counts)
```
